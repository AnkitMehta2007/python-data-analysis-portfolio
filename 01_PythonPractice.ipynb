# Data Type

 Data type batata hai data ka nature kya hai
 Example: number, text, collection


```python
# String = text data

name = "Ankit"
city = 'Delhi'

print(name)
print(type(name))
print(city)
print(type(city))
```

```python
text = "python"

print(text[0])     # p
print(text[-1])    # n
print(text[0:3])   # pyt
```

```python
msg = " data analysis "

print(msg.upper())
print(msg.lower())
print(msg.strip())
print(msg.replace("data", "python"))
print(msg.split())
```

```python
word = "Python123"

print(word.isalpha())
print(word.isdigit())
print(word.isalnum())
```

```python
# Integer = whole number
# Float = decimal number

age = 20
price = 99.75

print(age, type(age))
print(price, type(price))
```

```python
# Boolean = True or False

is_student = True
passed = False

print(is_student, type(is_student))
print(passed, type(passed))
```

```python
# List = ordered, changeable collection

marks = [70, 80, 90]
fruits = ["apple", "banana", "mango"]

print(marks)
print(type(marks))

marks[0] = 75
print(marks)
```

```python
nums = [10, 20, 30]
print(nums)
print(type(nums))
```

```python
print(nums[0])
print(nums[1:])
```

```python
nums.append(40)
nums.insert(1, 15)
nums.extend([50, 60])

print(nums)
```

```python
nums.remove(20)
popped = nums.pop()
print(nums)
print(popped)
```

```python
nums.sort()
nums.reverse()
print(nums)

print(len(nums))
print(nums.count(30))
```

Key points.
• Index based
• Mutable
• Most used in data analysis

```python
# Tuple = ordered, NOT changeable

dimensions = (1920, 1080)

print(dimensions)
print(type(dimensions))
```

```python
points = (10, 20, 30)
print(points)
print(type(points))
```

```python
print(points.count(20))
print(points.index(30))
```

Key points.
• Immutable

```python
# Set = unordered, unique values

numbers = {1, 2, 3, 3, 2}

print(numbers)
print(type(numbers))
```

```python
values = {1, 2, 3, 3, 2}
print(values)
print(type(values))
```

```python
values.add(4)
values.remove(2)
print(values)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

Key points.
• Duplicate allowed nahi

```python
# Dictionary = key value pair

student = {
    "name": "Ankit",
    "age": 20,
    "marks": 85
}

print(student)
print(type(student))
print(student["name"])
```

```python
print(student["name"])

student["marks"] = 90
student["city"] = "Delhi"
print(student)
```

```python
print(student.keys())
print(student.values())
print(student.items())
```

Key points.

#Mini Task

```python
# Create a list of student dictionaries
students = [
    {"name": "A", "marks": 70},
    {"name": "B", "marks": 85},
    {"name": "C", "marks": 40}
]

for s in students:
    if s["marks"] >= 40:
        print(s["name"], "Pass")
    else:
        print(s["name"], "Fail")
```
